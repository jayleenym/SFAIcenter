```python
import os, json
from tools import ONEDRIVE_PATH
essay_dir = os.path.join(ONEDRIVE_PATH, 'evaluation', 'eval_data', '9_multiple_to_essay', 'questions')

file = json.load(open(os.path.join(essay_dir, 'essay_questions_w_keyword_1st_서술형문제로변환.json'), 'r'))
```

```python
# keyword가 options나 explanation에 하나라도 있으면 OK, 둘 다 없으면 오류
errors = []

for item in file:
    keywords = [k.strip() for k in item.get('essay_keyword', '').split(',') if k.strip()]
    options_text = ' '.join(item.get('options', []))
    explanation_text = str(item.get('explanation', ''))
    
    # 각 keyword가 options나 explanation 둘 다 없는 경우만 오류
    missing_kw = [kw for kw in keywords if kw not in options_text and kw not in explanation_text]
    
    if missing_kw:
        errors.append((item, missing_kw))

print(f"오류: {len(errors)}개")
for item, missing_kw in errors:
    print(f"{'='*80}")
    print(f"ID: {item.get('file_id')}_{item.get('tag')}")
    print(f"둘 다 없는 keyword: {missing_kw}")
```
