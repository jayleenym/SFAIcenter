## abcd

```python
after_abcd = json.load(open(os.path.join(ONEDRIVE_PATH, 'evaluation/eval_data/7_multiple_rw/answer_type_classified.json')))
after_abcd_ids_raw = []
for m in after_abcd:
    if isinstance(m, dict):
        if 'question_id' in m.keys():
            after_abcd_ids_raw.append(m['question_id'])
        else:
            m['question_id'] = m['file_id'] + '_' + m['tag']
            after_abcd_ids_raw.append(m['question_id'])
    else:
        print(m)

print(f"전체 항목 수: {len(after_abcd_ids_raw)}")

# 중복 제거
after_abcd_ids = list(set(after_abcd_ids_raw))
print(f"중복 제거 후: {len(after_abcd_ids)}")
print(f"중복된 항목 수: {len(after_abcd_ids_raw) - len(after_abcd_ids)}")

# 중복된 question_id 찾기
from collections import Counter
id_counts = Counter(after_abcd_ids_raw)
duplicates = {id: count for id, count in id_counts.items() if count > 1}
print(f"\n중복된 question_id 개수: {len(duplicates)}")
if duplicates:
    print("\n중복된 question_id 예시 (상위 10개):")
    sorted_duplicates = sorted(duplicates.items(), key=lambda x: x[1], reverse=True)
    for qid, count in sorted_duplicates[:10]:
        print(f"  {qid}: {count}번 중복")
```

```python
# 중복된 question_id를 가진 실제 항목들 확인
if duplicates:
    # 첫 번째 중복된 question_id 예시 확인
    example_duplicate_id = sorted_duplicates[0][0]
    print(f"\n'{example_duplicate_id}'를 가진 모든 항목:")
    duplicate_items = [item for item in after_abcd if isinstance(item, dict) and item.get('question_id') == example_duplicate_id]
    for idx, item in enumerate(duplicate_items):
        print(f"\n--- 항목 {idx+1} ---")
        # 주요 키만 출력
        for key in ['question_id', 'question', 'answer', 'explanation']:
            if key in item:
                value = item[key]
                if isinstance(value, str) and len(value) > 100:
                    print(f"{key}: {value[:100]}...")
                else:
                    print(f"{key}: {value}")
```

```python
if duplicates:
    print(f"\n중복 제거 시작...")
    
    # question_id를 키로 하는 딕셔너리 생성 (나중에 추가된 항목이 이전 항목을 덮어씀)
    unique_items = {}
    duplicate_count = 0
    
    for item in after_abcd:
        if isinstance(item, dict) and 'question_id' in item:
            qid = item['question_id']
            if qid in unique_items:
                duplicate_count += 1
            unique_items[qid] = item  # 같은 ID가 있으면 덮어쓰기 (최신 유지)
        else:
            # question_id가 없는 항목은 그대로 유지
            unique_items[id(item)] = item
    
    # 딕셔너리 값을 리스트로 변환
    deduplicated_after_abcd = list(unique_items.values())
    
    print(f"원본 항목 수: {len(after_abcd)}")
    print(f"중복 제거 후: {len(deduplicated_after_abcd)}")
    print(f"제거된 중복 항목 수: {duplicate_count}")
    
    # 파일에 저장 (백업 후)
    result_file = os.path.join(ONEDRIVE_PATH, 'evaluation/eval_data/7_multiple_rw/pick_abcd/result.json')
    backup_file = os.path.join(ONEDRIVE_PATH, 'evaluation/eval_data/7_multiple_rw/pick_abcd/result.json.backup')
    
    # 백업 생성
    import shutil
    if os.path.exists(result_file):
        shutil.copy2(result_file, backup_file)
        print(f"\n백업 파일 생성: {backup_file}")
    
    # 중복 제거된 데이터 저장
    with open(result_file, 'w', encoding='utf-8') as f:
        json.dump(deduplicated_after_abcd, f, ensure_ascii=False, indent=4)
    
    print(f"중복 제거된 데이터 저장 완료: {result_file}")
    
    # 업데이트된 after_abcd 사용
    after_abcd = deduplicated_after_abcd
    after_abcd_ids_raw = [item['question_id'] for item in after_abcd if isinstance(item, dict) and 'question_id' in item]
    after_abcd_ids = list(set(after_abcd_ids_raw))
    print(f"\n최종 통계:")
    print(f"  전체 항목 수: {len(after_abcd_ids_raw)}")
    print(f"  고유 question_id 수: {len(after_abcd_ids)}")
else:
    print("중복이 없습니다.")
```

### 결과 파일 포맷 변경

```python
import time
from tools.core.llm_query import LLMQuery
llmquery = LLMQuery()


def is_essentially_same(llm, text1, text2):
    system_prompt = """아래 question과 original_question을 비교하고, 차이를 확인하세요. 최대한 original_question을 사용하지만, question이 original_question과 문장이 동일하고 오타 및 띄어쓰기가 교정된 버전이라면 question을 선택하세요. 차이가 없다면 original_question을 선택하세요.
절대로 내용을 출력하지 말고, question인지 original_question인지만 출력하세요.

출력형식
question / original_question"""
    user_prompt = f"""
question: {text1},
original_question: {text2}
"""
    response = llm.query_openrouter(system_prompt, user_prompt, model_name='google/gemini-2.5-flash')
    return response

new_abcd = []
for original in original_abcd:
    for after in after_abcd:
        if after['question_id'] == original['file_id']+ '_' + original['tag'] and original['answer_type'] == 'abcd':
            new_after = {
                'file_id': original['file_id'],
                'title': original['title'],
                'chapter': original['chapter'],
                'tag': original['tag'],
                'domain': original['domain'],
                'subdomain': original['subdomain'],
                'classification_reason': original['classification_reason'],
                'is_calculation': original['is_calculation'],
                'answer_type': original['answer_type'],
                'question': after['question'],
                'original_question': original['question'],
                # 'original_options': original['options'],
                # 'original_answer': original['answer'],
                'options': after['options'],
                'answer': after['answer'],
                'explanation': after['explanation'],
                'original_explanation': original['explanation']
            }
            # 공백/줄바꿈 등 사소한 차이만 있으면 question 사용 (original_question 제거)
            # 실제 내용 차이가 있으면 original_question 유지
            llm_choice = is_essentially_same(llmquery, new_after['question'], new_after['original_question'])
            if llm_choice == 'question':
                new_after.pop('original_question')
            elif llm_choice == 'original_question':
                new_after['question'] = new_after['original_question']
                new_after.pop('original_question')
            else:
                print(llm_choice)
                print(new_after)
            
            new_abcd.append(new_after)
```

# 질문이 이상함 - 아무것도 없이 '옳은 것을 고르시오'

```python
import os, json
from tqdm import tqdm
from tools.pipeline.config import ONEDRIVE_PATH
from tools.core.llm_query import LLMQuery

# LLM 초기화 및 함수 정의
llmquery = LLMQuery()

def is_useful_question(llm, text1, text2):
    system_prompt = """
작업:
- 주어진 question과 original_question을 비교해 '새로운 질문' 한 줄을 생성한다.

규칙:
1) question과 original_question을 비교해
   - 주제(무엇에 대한 질문인지)
   - 판단 방향(옳은 것 vs. 옳지 않은 것, 적절한 vs. 부적절한 등)을 파악한다.
2) 최종 판단 방향은 question을 따른다. question에 방향이 없으면 original_question을 따른다.
3) question에 주제가 없거나 불명확하면, original_question에서 주제만 추출해 question의 형식/방향과 결합한다.
   - 주제 추출 시 '다음 중/설명 중/가장/옳은/옳지 않은/적절(하지) 않은/맞는/틀린/해당하지 않는' 등 판단 표현은 제거하고, 대상/맥락만 남긴다.
4) question에 주제와 판단 방향이 모두 명확하면 question을 그대로 사용한다.
5) 어투와 종결('…고르시오.', '…것은?' 등)은 original_question의 형식을 최대한 유지한다.
6) 의미 왜곡 없이 자연스러운 한국어로 다듬되, 불필요한 수식어는 추가하지 않는다.
7) 출력 형식은 한 줄로만: 새로운 질문: {문장}

예시 입력:
- question: 옳은 것을 모두 고르시오.
- original_question: 고객과 생산적인 대화를 위한 방법으로 다음 설명 중 가장 적절하지 않은 것은?

예시 출력:
새로운 질문: 고객과 생산적인 대화를 위한 방법으로 다음 설명 중 옳은 것을 모두 고르시오."""
    user_prompt = f"""
===================
입력:
question: {text1},
original_question: {text2}
"""
    response = llm.query_openrouter(system_prompt, user_prompt, model_name='google/gemini-2.5-flash')
    return response

# 데이터 로드
multiple = json.load(open(os.path.join(ONEDRIVE_PATH, 'evaluation', 'eval_data', '7_multiple_rw', 'answer_type_classified.json')))
original_questions = [m for m in multiple if m['answer_type'] == 'right']

# original_questions를 딕셔너리로 변환하여 빠른 조회 (성능 개선)
original_dict = {f"{m['file_id']}_{m['tag']}": m for m in original_questions if m.get('answer_type') == 'right'}

# pick_right 폴더 내의 숫자 폴더들 처리
pick_dir = os.path.join(ONEDRIVE_PATH, 'evaluation', 'eval_data', '7_multiple_rw', 'pick_right')
for number in sorted(os.listdir(pick_dir)):  # sorted로 2, 3, 4, 5 순서 보장
    # 디렉토리인지 확인
    number_path = os.path.join(pick_dir, number)
    if not os.path.isdir(number_path):
        continue
    
    print('pick_right', number)
    file_name = os.path.join(number_path, 'result.json')
    
    # 파일 존재 확인
    if not os.path.exists(file_name):
        print(f"  경고: {file_name} 파일이 존재하지 않습니다. 건너뜁니다.")
        continue
    
    try:
        after_questions = json.load(open(file_name, 'r', encoding='utf-8'))
    except Exception as e:
        print(f"  오류: {file_name} 파일을 읽는 중 오류 발생: {e}")
        continue

    new_questions = []
    for after in tqdm(after_questions, desc=f"처리 중 ({number})"):
        # question_id로 빠르게 조회
        question_id = after.get('question_id')
        if not question_id:
            print(f"  경고: question_id가 없는 항목 발견: {after}")
            continue
        
        original = original_dict.get(question_id)
        if original and original.get('answer_type') == 'right':
            new_after = {
                'file_id': original['file_id'],
                'title': original['title'],
                'chapter': original['chapter'],
                'tag': original['tag'],
                'domain': original['domain'],
                'subdomain': original['subdomain'],
                # 'classification_reason': original['classification_reason'],
                'is_calculation': original['is_calculation'],
                'answer_type': original['answer_type'],
                'question': after['question'],
                'original_question': original['question'],
                # 'original_options': original['options'],
                # 'original_answer': original['answer'],
                'options': after['options'],
                'answer': after['answer'],
                'explanation': after['explanation'],
                'original_explanation': original['explanation']
            }
            # 공백/줄바꿈 등 사소한 차이만 있으면 question 사용 (original_question 제거)
            # 실제 내용 차이가 있으면 original_question 유지
            try:
                llm_question = is_useful_question(llmquery, new_after['question'], new_after['original_question'])
                new_after['question'] = llm_question.replace('새로운 질문: ', '').strip()
            except Exception as e:
                print(f"  경고: LLM 처리 중 오류 발생 (question_id: {question_id}): {e}")
                # 오류 발생 시 원본 question 유지
                pass
            
            new_questions.append(new_after)
        elif not original:
            print(f"  경고: original_questions에서 매칭되지 않은 question_id: {question_id}")

    # 백업 생성
    if os.path.exists(file_name):
        backup_file = file_name + '.bak'
        try:
            with open(backup_file, 'w', encoding='utf-8') as f:
                json.dump(after_questions, f, ensure_ascii=False, indent=4)
            print(f"  백업 생성: {backup_file}")
        except Exception as e:
            print(f"  경고: 백업 생성 중 오류 발생: {e}")

    # 새 파일 저장
    try:
        with open(file_name, 'w', encoding='utf-8') as f:
            json.dump(new_questions, f, ensure_ascii=False, indent=4)
        print(f"  저장 완료: {file_name} (총 {len(new_questions)}개 항목)")
    except Exception as e:
        print(f"  오류: 파일 저장 중 오류 발생: {e}")
```
