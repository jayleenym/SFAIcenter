```python
# 배치 처리 버전 (더 효율적, 선택사항)
# 위의 하나씩 처리하는 대신 배치로 처리하려면 아래 코드를 사용하세요

def classify_batch(questions, batch_size=10):
    """배치 단위로 문제를 분류"""
    failed_items = []
    
    for i in tqdm(range(0, len(questions), batch_size), desc="배치 분류 중"):
        batch = questions[i:i + batch_size]
        
        try:
            # 배치용 user_prompt 생성
            user_prompt = ""
            for item in batch:
                user_prompt += f"""문제 ID: {item.get('file_id', '')}_{item.get('tag', '')}
질문: {item.get('question', '')}
선택지: {', '.join(item.get('options', []))}
정답: {item.get('answer', '')}
====================
"""
            
            # LLM 호출
            response = llm.query_openrouter(system_prompt, user_prompt)
            
            # 응답을 줄 단위로 분리 (각 문제당 하나의 답변)
            responses = [r.strip().lower() for r in response.split('\n') if r.strip()]
            
            # 각 문제에 answer_type 할당
            for j, item in enumerate(batch):
                if j < len(responses):
                    answer_type = responses[j]
                    
                    # 유효한 값인지 확인
                    if answer_type not in ['right', 'wrong', 'abcd']:
                        if 'right' in answer_type:
                            answer_type = 'right'
                        elif 'wrong' in answer_type:
                            answer_type = 'wrong'
                        elif 'abcd' in answer_type:
                            answer_type = 'abcd'
                        else:
                            answer_type = 'right'
                            failed_items.append({
                                'index': i + j,
                                'item_id': f"{item.get('file_id', '')}_{item.get('tag', '')}",
                                'response': answer_type,
                                'question': item.get('question', '')[:100]
                            })
                    
                    item['answer_type'] = answer_type
                else:
                    item['answer_type'] = 'error'
                    failed_items.append({
                        'index': i + j,
                        'item_id': f"{item.get('file_id', '')}_{item.get('tag', '')}",
                        'error': '응답 개수 부족',
                        'question': item.get('question', '')[:100]
                    })
            
            # API 호출 간격 조절
            time.sleep(1.2)
            
        except Exception as e:
            print(f"배치 오류 발생 (인덱스 {i}): {e}")
            for j, item in enumerate(batch):
                item['answer_type'] = 'error'
                failed_items.append({
                    'index': i + j,
                    'item_id': f"{item.get('file_id', '')}_{item.get('tag', '')}",
                    'error': str(e),
                    'question': item.get('question', '')[:100]
                })
    
    return failed_items

# 배치 처리 실행 (주석 해제하여 사용)
# failed_items = classify_batch(multiple, batch_size=10)
# print(f"\n배치 분류 완료!")
# print(f"전체: {len(multiple)}")
# print(f"실패: {len(failed_items)}")
```
