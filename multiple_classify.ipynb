```python
# AnswerTypeClassifier를 사용하여 각 문제를 분류하고 answer_type 키 추가
from tools.qna.processing import AnswerTypeClassifier

# 분류기 초기화
classifier = AnswerTypeClassifier()

# multiple 리스트의 각 문제에 answer_type 키 추가
# 실패한 항목은 answer_type이 "" (빈 문자열)로 설정됨
results = classifier.process_all_questions(
    questions=multiple,
    batch_size=10,
    model='openai/gpt-5'
)

# 결과는 multiple 리스트에 직접 반영됨 (answer_type 키가 추가됨)
# 결과는 evaluation/eval_data/7_multiple_rw/answer_type_classified.json에 저장됨
# 로그는 logs/multiple_answer_classify.log에 저장됨

print(f"\n분류 완료!")
print(f"전체: {len(multiple)}")
print(f"결과 저장 위치: evaluation/eval_data/7_multiple_rw/answer_type_classified_gpt5.json")
print(f"로그 저장 위치: logs/multiple_answer_classify_gpt5.log")
```

```python
# 분류 결과 통계 확인
answer_type_counts = {}
for item in multiple:
    answer_type = item.get('answer_type', 'unknown')
    # 빈 문자열인 경우 "(실패)"로 표시
    display_type = answer_type if answer_type else "(실패)"
    answer_type_counts[display_type] = answer_type_counts.get(display_type, 0) + 1

print("answer_type 분류 결과:")
for answer_type, count in sorted(answer_type_counts.items()):
    print(f"  {answer_type}: {count}")
```
